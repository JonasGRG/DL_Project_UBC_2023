```python
import pandas as pd
```

```python
# Load the CSV file into a pandas DataFrame
df = pd.read_csv('data/results-imagenet.csv')
df['img_size'] = pd.to_numeric(df['img_size'], errors='coerce')
df['img_size'] = pd.to_numeric(df['img_size'], errors='coerce')
df['param_count'] = pd.to_numeric(df['param_count'], errors='coerce')

# Filter the DataFrame for models with fewer than 100 million parameters and image size of 224
filtered_df_100 = df[(df['param_count'] < 100) & (df['img_size'] == 224)]
filtered_df_50 = df[(df['param_count'] < 50) & (df['img_size'] == 224)]
filtered_df_25 = df[(df['param_count'] < 25) & (df['img_size'] == 224)]


# Sort the filtered DataFrame by 'top1' accuracy in descending order and select the top 10 models
top_models_below_100 = filtered_df_100.sort_values(by='top1', ascending=False).head(5)
top_models_below_50 = filtered_df_50.sort_values(by='top1', ascending=False).head(5)
top_models_below_25 = filtered_df_25.sort_values(by='top1', ascending=False).head(5)

# Display the top 10 models
top_models_below_100
```
